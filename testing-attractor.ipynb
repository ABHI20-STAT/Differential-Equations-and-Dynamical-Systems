```python
import numpy as np
import matplotlib.pyplot as plt

def lorenz(n=20000, dt=0.01, sigma=10.0, rho=28.0, beta=8/3, x0=(1.0, 1.0, 1.0)):
    x = np.empty(n); y = np.empty(n); z = np.empty(n)
    x[0], y[0], z[0] = x0
    for i in range(n - 1):
        dx = sigma * (y[i] - x[i])
        dy = x[i] * (rho - z[i]) - y[i]
        dz = x[i] * y[i] - beta * z[i]
        x[i+1] = x[i] + dt * dx
        y[i+1] = y[i] + dt * dy
        z[i+1] = z[i] + dt * dz
    return x, y, z

x, y, z = lorenz()

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot(x, y, z, linewidth=0.3)
ax.set_title("Lorenz Attractor")
ax.set_xlabel("x"); ax.set_ylabel("y"); ax.set_zlabel("z")
plt.show()


import numpy as np
import matplotlib.pyplot as plt

def clifford(n=2_000_00, a=-1.4, b=1.6, c=1.0, d=0.7, x0=0.1, y0=0.1):
    x = np.empty(n); y = np.empty(n)
    x[0], y[0] = x0, y0
    for i in range(n - 1):
        x[i+1] = np.sin(a * y[i]) + c * np.cos(a * x[i])
        y[i+1] = np.sin(b * x[i]) + d * np.cos(b * y[i])
    return x, y

x, y = clifford()

plt.figure()
plt.scatter(x, y, s=0.05, alpha=0.6)
plt.title("Clifford Attractor")
plt.axis("equal"); plt.axis("off")
plt.show()

```
